```python
#question4
def list_mean(input_list):
    k = 0
    count = 0
    for i in input_list:
        k = i + k
        count += 1
        mean = k/count
    return mean

given_list = [10, 56.7, 56, 89, 100, 99, 87.5, 34, 985, 10] 
print(list_mean(given_list))
```

```python
#question6
from abc import ABC, abstractmethod
from math import tan, pi

class Shape(ABC):
    @abstractmethod
    def area(self):
        pass
    
class Circle(Shape):
    def area(self,radius):
        self.radius = radius
        circle_area = pi*(self.radius**2)
        return circle_area
class FilledCircle(Circle):
    super(Circle)
class Dot(FilledCircle):
    super(FilledCircle)

class Rectangle(Shape):
    def area(self,length,breadth):
        self.length = length
        self.breadth = breadth
        rectangle_area = self.length*self.breadth
        return rectangle_area

class Polygon(Shape):
    def area(self,no_of_sides,side_length):
        self.no_of_sides = no_of_sides
        self.side_length = side_length
        polygon_area = (no_of_sides*(side_length**2))/(4*tan(pi/no_of_sides))
        return polygon_area
class Square(Polygon):
    super(Polygon)
class FilledPolygon(Polygon):
    super(Polygon)
    
#areas_calculation

circle = Circle()
Area_of_circle = circle.area(20)
print('Area of Circle:', Area_of_circle)

rectangle = Rectangle()
Area_of_rectagle = rectangle.area(15,10)
print('Area of Rectagle:', Area_of_rectagle)

polygon = Polygon()
Area_of_polygon = polygon.area(8,3)
print('Area of Polygon:', Area_of_polygon)

filledcircle = FilledCircle()
Area_of_filledcircle = filledcircle.area(20)
print('Area of FilledCircle:', Area_of_filledcircle)

dot = Dot()
Area_of_dot = dot.area(0.00001)
print('Area of Dot:', Area_of_dot)

square = Square()
Area_of_square = square.area(4,4)
print('Area of Square:', Area_of_square)

filledpolygon = FilledPolygon()
Area_of_filledpolygon = filledpolygon.area(8,3)
print('Area of FilledPolygon:', Area_of_filledpolygon)
```
