```python
#question5
import matplotlib.pyplot as plt
plt.close()
principal = 5000
interestRate = 0.15
years = 10
simple_values = []
i = -1
for s in range(years+1):
    i = i+1
    amount = principal + i*principal*interestRate
    simple_values.append(amount)
comp_values = []
for i in range(years+1):
    comp_values.append(principal)
    principal += principal*interestRate
plt.plot(simple_values)
plt.plot(comp_values)
plt.title('Simple Interest vs Compound Interest')
plt.xlabel('Years')
plt.legend(['Simple Interest','Compound Interest'])
plt.ylabel('Amount')
plt.show()
```
